```python
import json
import tweepy
import asyncio
from Constants import api

errors = []

# function requests tweets by tweet ids found in other dataframe from Twitter API
# API authentication was done in other file (Constants.py) to hide keys and tokens 

async def get_tweets():
    index = 0
    tweet_index = 0
    
    with open('tweets_json.txt', 'w') as outfile:

        while index < 2076:
            
        # uses asyncio to wait 15 minutes between each request due to twitter api rate limits
            if index == 900: 
                print('entered 900 sleep')
                await asyncio.sleep(960) 
                
            if index == 1800:
                print('entered 1800 sleep')
                await asyncio.sleep(900)
          
        # in a try/catch to send ids to an errors list in case any ids have been deleted and can no longer be retrieved 
            try: 
                tweets_to_get = tweet_ids['tweet_id'][tweet_index]
                tweets = api.get_status(tweets_to_get, tweet_mode='extended')
                
        # writes responses in json format in a txt file        
                json.dump(tweets._json, outfile)
                outfile.write('\n')
                print("Request successful!")
                tweet_index += 1
                index += 1
                
        # creates a list of errors to be referenced later on to know which id's from the dataframes were unavailable
            except tweepy.TweepError as err:
                errors.append(tweets_to_get)
                tweet_index += 1
                index += 1
                print("ID not found")
                
            pass
```

## <span style="color:red"> STOP -</span>

#### <span style="color:red"> Cell below takes approx 30 min to run - do not rerun if tweets_json.txt is already available. </span>

```python
# asyncio typically runs into errors in jupyter notebook due to the kernel continuously running 
# big shoutout to Vinicuis Fortuna on GitHub providing this workaround 
# https://github.com/ipython/ipython/issues/11030#issuecomment-377723960

asyncio.run_coroutine_threadsafe(get_tweets(), asyncio.get_event_loop())
```

Is there a correlation between how many times a tweet was favorited and how many times a tweet was retweeted?

```python
from scipy.stats import pearsonr
```

```python
# finds the correlation coefficient and p-value of the retweet count and favorite count
# the closer the first number in the returned tuple is to 1 or -1, the stronger the correlation
# the second number, the p-value, tells us the correlation is determined to be statistically signficant if it is less than 0.05

pearsonr(df_retweets_favorites['retweet_count'], df_retweets_favorites['favorite_count'])
```

As seen in the scatter plot and the high correlation coefficient (0.88221708) above, there is a strong positive correlation between how many times a tweet was favorited and how many times a tweet was retweeted.
