```python
import json
import tweepy
import asyncio
from Constants import api

errors = []

# function requests tweets by tweet ids found in other dataframe from Twitter API
# API authentication was done in other file (Constants.py) to hide keys and tokens 

async def get_tweets():
    index = 0
    tweet_index = 0
    
    with open('tweets_json.txt', 'w') as outfile:

        while index < 2076:
            
        # uses asyncio to wait 15 minutes between each request due to twitter api rate limits
            if index == 900: 
                print('entered 900 sleep')
                await asyncio.sleep(960) 
                
            if index == 1800:
                print('entered 1800 sleep')
                await asyncio.sleep(900)
          
        # in a try/catch to send ids to an errors list in case any ids have been deleted and can no longer be retrieved 
            try: 
                tweets_to_get = tweet_ids['tweet_id'][tweet_index]
                tweets = api.get_status(tweets_to_get, tweet_mode='extended')
                
        # writes responses in json format in a txt file        
                json.dump(tweets._json, outfile)
                outfile.write('\n')
                print("Request successful!")
                tweet_index += 1
                index += 1
                
        # creates a list of errors to be referenced later on to know which id's from the dataframes were unavailable
            except tweepy.TweepError as err:
                errors.append(tweets_to_get)
                tweet_index += 1
                index += 1
                print("ID not found")
                
            pass
```

## <span style="color:red"> STOP -</span>

#### <span style="color:red"> Cell below takes approx 30 min to run - do not rerun if tweets_json.txt is already available. </span>

```python
# asyncio typically runs into errors in jupyter notebook due to the kernel continuously running 
# big shoutout to Vinicuis Fortuna on GitHub providing this workaround 
# https://github.com/ipython/ipython/issues/11030#issuecomment-377723960

asyncio.run_coroutine_threadsafe(get_tweets(), asyncio.get_event_loop())
```

```python
# reads txt file

with open('tweets_json.txt') as file:
    json_tweets = []
    for line in file:
        json_tweets.append(line)
```

```python
# extracts tweet_id, retweet count, and favorite count from the json objects in txt file

tweet_list = []
for tweet in json_tweets:
    tweet_id = json.loads(tweet)['id']
    retweet_count = json.loads(tweet)['retweet_count']
    favorite_count = json.loads(tweet)['favorite_count']
    tweet_list.append({'tweet_id': tweet_id, 'retweet_count': retweet_count, 'favorite_count':favorite_count})
```

```python
# creates dataframe from json objects extracted above

api_tweets_df = pd.DataFrame(tweet_list, columns = ['tweet_id', 'retweet_count', 'favorite_count'])
```

#### Quality Issue #4

#### Define

3.	Remove ids missing from "api_tweets_df" from "df_image_predictions." 
        - there were ids missing from the api request results which need to be deleted from df_image_predictions so the user_ids all match

#### Code

```python
ids_to_delete
```

```python
# loop to get row index where tweet_id == tweet_ids not found in api request

row_indices_list = []

for id in ids_to_delete:

    index = clean_image_predictions.index
    condition = clean_image_predictions['tweet_id'] == id
    row_indices = index[condition].tolist()

    row_indices_list.append(row_indices)
```

```python
# found one line solution to flatten row_indices list into one list on stackoverflow - https://stackoverflow.com/questions/952914/how-to-make-a-flat-list-out-of-a-list-of-lists
row_indices_list = [item for sublist in row_indices_list for item in sublist]
```

```python
# checking row_indices_list to make sure it is a "flat list" (instead of a list of lists)
row_indices_list
```

```python
# loop through list and drop rows with index where tweet_id == tweet_ids not found from api requests
for index in row_indices_list:
    clean_image_predictions.drop(index={index}, inplace=True)
```

```python
clean_image_predictions.reset_index(inplace=True)
```

```python
clean_image_predictions.drop(columns={'index'}, inplace=True)
```
